```python
import wikipedia
import wikipediaapi
```

```python
import pandas as pd
df = pd.read_excel(r"<data_path>\user_names.xlsx")
print(df.shape)
print(df.head())
```

```python
userid = []
editcount = []
gender = []
blockinfo = []
emailable = []
```

```python
for i in range(1,76):
    a = df["Usernames"][(i-1)*49:i*49]
    U = ""
    for n in range(len(a)):
        b = a.values[n]
        U = U + str(b) + "|"
    
    import requests
    S = requests.Session()

    URL = "https://en.wikipedia.org/w/api.php"

    PARAMS = {
        "action": "query",
        "format": "json",
        "list": "users",
        "ususers": str(U),
        "usprop": "blockinfo|editcount|emailable|gender"
    }

    R = S.get(url=URL, params=PARAMS)
    DATA = R.json()
    DATA["query"]["users"].pop(0)

    USERS = DATA["query"]["users"]
    
    for u in USERS:
        if 'userid' in u:
            userid.append(u["userid"])
        else:
            userid.append("unknown")
        if 'editcount' in u:
            editcount.append(u["editcount"])
        else:
            editcount.append("unknown")
        if 'gender' in u:
            gender.append(u["gender"])
        else:
            gender.append("unknown")
        if 'emailable' in u:
            emailable.append(1)
        else:
            emailable.append(0)
        if 'blockid' in u:
            blockinfo.append(1)
        else:
            blockinfo.append(0)
    
    for j in range((i-1)*49,i*49):
        df.loc[j, "User ID"] = int(userid[j])
        df.loc[j,'Edit  Count'] = int(editcount[j])
        df.loc[j,'Gender'] = gender[j]
        df.loc[j,'Blockinfo'] = int(blockinfo[j])
        df.loc[j,'Emailable'] = int(emailable[j])
```

```python
for i in range(1,2):
    a = df["Usernames"][3675:3700]
    U = ""
    for n in range(len(a)):
        b = a.values[n]
        U = U + str(b) + "|"
    
    import requests
    S = requests.Session()

    URL = "https://en.wikipedia.org/w/api.php"

    PARAMS = {
        "action": "query",
        "format": "json",
        "list": "users",
        "ususers": str(U),
        "usprop": "blockinfo|editcount|emailable|gender"
    }

    R = S.get(url=URL, params=PARAMS)
    DATA = R.json()
    DATA["query"]["users"].pop(0)

    USERS = DATA["query"]["users"]
    
    for u in USERS:
        if 'userid' in u:
            userid.append(u["userid"])
        else:
            userid.append("unknown")
        if 'editcount' in u:
            editcount.append(u["editcount"])
        else:
            editcount.append("unknown")
        if 'gender' in u:
            gender.append(u["gender"])
        else:
            gender.append("unknown")
        if 'emailable' in u:
            emailable.append(1)
        else:
            emailable.append(0)
        if 'blockid' in u:
            blockinfo.append(1)
        else:
            blockinfo.append(0)
    
    for j in range(3675,3700):
        df.loc[j, "User ID"] = int(userid[j])
        df.loc[j,'Edit  Count'] = int(editcount[j])
        df.loc[j,'Gender'] = gender[j]
        df.loc[j,'Blockinfo'] = int(blockinfo[j])
        df.loc[j,'Emailable'] = int(emailable[j])
```

```python
for i in range(1,3700):
    a = df["Usernames"][i]
    try:
        df.loc[i, "About"] = wikipedia.summary("User:"+str(a))
    except:
        df.loc[i, "About"] = "unknown"
```
